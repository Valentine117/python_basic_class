# 도형들 1

인자 `x`를 입력받아 저장하고 `calcarea` 메소드로 넓이를 출력하는 정삼각형 클래스 `RegularTriangle`, 정사각형 클래스 `Square`와 원 클래스 `Circle`을 만들자.
<br>

### hint
- 한 변의 길이가 $x$인 정삼각형, 정사각형의 넓이는 각각 ${\sqrt{3} \over 4} x^2$, $x^2$이고 반지름이 $x$인 원의 넓이는 $\pi x^2$이다.
- 세 클래스 모두 생성자(`__init__`)에서 인자 `x`를 입력받아 `self.x`에 저장한다.
- `math` 모듈을 이용하면 `pi`와 `sqrt`(제곱근)를 사용할 수 있다.
```python
>>> from math import pi, sqrt
>>> pi
3.141592653589793
>>> sqrt(3)
1.7320508075688772
```

<br>

---

목표 : 

다음을 수행

```python

t.calcarea()
s.calcarea()
c.calcarea()
```

```

```

# 공격 & 방어 (★★)

만들어야 할 클래스들의 상세 정보가 적혀있는 문서를 명세서라 한다. 클래스들의 명세들은 각 클래스와 메소드의 docstring으로 표시해놨다. docstring은 클래스나 메소드의 설명을 적을 수 있으며 실제 작동에 영향을 주지 않는다.  
<br>

### hint
- `pass`라고 써있는 부분만 수정한다. (인자와 메소드를 추가하거나 삭제하면 안된다.)
- 물리 공격을 적에게 가하면 적은 (내 물리 공격력) - (적 물리 방어력) 만큼 HP가 빠진다.
- 마법 공격을 적에게 가하면 적은 (내 마법 공격력) - (적 마법 방어력) 만큼 HP가 빠진다.
- 만약 (공격력) - (방어력) 값이 음수라면 데미지는 0이다.
- HP가 0이 되면 사망한다. 
- 공격받을 때마다 누구에게 얼마만큼 피해를 입었는지 출력한다.
- 공격받다 사망하면 누구에게 공격받아 사망했다고 출력한다.
<br>

---

목표 : 

다음을 수행

```python
worrior = Worrior('발렌타인', 2283, 327, 624, 27, 453)
magician = Magician('하우론', 1623, 133, 256, 1024, 556)

worrior.attack_physically(magician)

magician.attack_magically(worrior)

magician.attack_magically(worrior)
```

출력  
```

발렌타인이(가) 하우론에게 공격받음.
받은 데미지 : 0, 남은 HP : 570

플레이어 발렌타인이(가) 하우론에게 공격받아 사망했습니다.

```

````python
class Classes:
    '''
    객체의 모든 직업의 공통적인 행동을 관리하는 클래스이다.
    모든 직업클래스들은 이 클래스를 상속받는다.
    멤버변수:
        class_name : 객체의 직업명이다. worrior또는 maigician을 가진다.
        player_name : 객체를 소유한 플레이어의 이름이다.
        HP : 현재 남은 HP의 양이다.
        physical_attack : 물리 공격력
        physical_armor : 물리 방어력
        magical_attack : 마법 공격력
        magical_armor : 마법 방어력
    '''
    def __init__(
        self, class_name, player_name,
        maxHP,
        physical_attack, physical_armor,
        magical_attack, magical_armor):
        '''
        멤버변수들을 초기화한다.
        인자:
            class_name : 직업명이다. worrior또는 maigician을 가진다.
            player_name : 플레이어 이름이다.
            maxHP : 최대 HP의 양이다.
            physical_attack : 물리 공격력
            physical_armor : 물리 방어력
            magical_attack : 마법 공격력
            magical_armor : 마법 방어력
        '''
        pass
        
    def attack_physically(self, enemy):
        '''
        적에게 물리 공격을 가한다.
        인자 :
             enemy : 때릴 적 객체이다.
        '''
        enemy.loseHP(self, self.physical_attack - enemy.physical_armor)
        
    def attack_magically(self, enemy):
        '''
        적에게 마법 공격을 가한다.
        인자 :
             enemy : 때릴 적 객체이다.
        '''
        enemy.loseHP(self, self.magical_attack - enemy.magical_armor)
        
    def loseHP(self, enemy, amount):
        '''
        데미지를 계산해 이 객체(self)의 HP를 깎고 공격받은 정보를 출력한다.
        이 객체가 죽었다면 사망했다고 표시한다.
        정확한 표시 방법은 출력 예를 참조한다.
        인자 :
             enemy : 때린 적 객체이다.
             amount : (적의 공격력 - 내 방어력) 값을 넘겨 받는다. 이 값이 음수라면 받은 데미지는 0이다. 
        '''
        pass

        
class Worrior(Classes):
    
    def __init__(self, player_name, maxHP, 
        physical_attack, physical_armor,
        magical_attack, magical_armor):
        '''
        super함수로 Classes의 생성자를 호출한다.
        그 이외의 기능은 없다.
        '''
        pass
        
        
class Magician(Classes):
    
    def __init__(self, player_name, maxHP,
        physical_attack, physical_armor,
        magical_attack, magical_armor):
        '''
        super함수로 Classes의 생성자를 호출한다.
        그 이외의 기능은 없다.
        '''
        pass
````

````python
worrior = Worrior('발렌타인', 2283, 327, 624, 27, 453)
magician = Magician('하우론', 1623, 133, 256, 1024, 556)

magician.attack_magically(worrior)
worrior.attack_physically(magician)
magician.attack_magically(worrior)
magician.attack_magically(worrior)
magician.attack_physically(worrior)
magician.attack_magically(worrior)
````
