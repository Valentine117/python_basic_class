# 페이스북 게시물

페이스북 타임라인에서 보여지는 게시물 1개 안에는 다음과 같은 내용들이 있다.
- 게시자 이름
- 올린 시간
- 공개 범위 (전체 공개, 친구만 공개 등...)
- 내용
- 좋아요 한 사람들
- 댓글 단 사람들
- 공유한 사람들
<br>

이를 이용해 다음을 수행하는 클래스 `FacebookPost`를 만들어보자.
<br>

### hint
- 파이썬의 클래스 생성자 (`__init__`)을 이용하자!
- `__dict__`는 객체의 원소를 보여주는 파이썬 빌트인 변수이다. 이 문제에서는 값이 잘 대입되었는지 확인하는 용도로만 사용되었다. 몰라도 된다!

<br>

---

목표 : 

다음을 수행

```python
post = FacebookPost(
    poster='Bad-Mouth',
    time='2시간',
    public=True,
    content='오토바이 : ????.gif\n존나 황당하겠다 ㄹㅇㅋㅋㅋㅋ',
    wholiked=['김민석', '유재건', '서용훈', '박성준'],
    whocommented=['김성재', '조영준', '우수한'],
    whoshared=['김연지'])
print(post.__dict__)
```

출력  
```
{'poster': 'Bad-Mouth',
 'time': '2시간',
 'content': '오토바이 : ????.gif\n존나 황당하겠다 ㄹㅇㅋㅋㅋㅋ',
 'wholiked': ['김민석', '유재건', '서용훈', '박성준'],
 'whocommented': ['김성재', '조영준', '우수한'],
 'whoshared': ['김연지']}```

````python
# Sol 페이스북 게시물
class FacebookPost:
    def __init__(
        self, poster, time, public,
    content, wholiked, whocommented,
    whoshared):
        self.poster = poster
        self.time = time
        self.content = content
        self.wholiked = wholiked
        self.whocommented = whocommented
        self.whoshared = whoshared
````

````python
post = FacebookPost(
    poster='Bad-Mouth',
    time='2시간',
    public=True,
    content='오토바이 : ????.gif\n존나 황당하겠다 ㄹㅇㅋㅋㅋㅋ',
    wholiked=['김민석', '유재건', '서용훈', '박성준'],
    whocommented=['김성재', '조영준', '우수한'],
    whoshared=['김연지'])
print(post.__dict__)
````

# 게시물 보여주기 (★)

위의 예제에서는 값을 대입하기까지만 했지만 대입한 값을 형식에 맞게 보여줄 수 있도록 ***클래스를 재정의해보자!***
<br>

### hint
- 클래스를 통째로 재정의 하세요!
- 보여줄 때는 show 메소드를 이용합니다.
- 살짝 고민해야 하는 부분이 있습니다 잘 생각해보세요.

<br>

---

목표 : 

다음을 수행

```python
post = FacebookPost(
    poster='Bad-Mouth',
    time='2시간',
    public=True,
    content='오토바이 : ????.gif\n존나 황당하겠다 ㄹㅇㅋㅋㅋㅋ',
    wholiked=['김민석', '유재건', '서용훈', '박성준'],
    whocommented=['김성재', '조영준', '우수한'],
    whoshared=['김연지'])
post.show()
```

출력  
```
게시자 : Bad-Mouth

올린 시간 : 2시간

내용 : 
오토바이 : ????.gif
존나 황당하겠다 ㄹㅇㅋㅋㅋㅋ

좋아요 한 사람 : 
김민석
유재건
서용훈
박성준

댓글 남긴 사람 : 
김성재
조영준
우수한

공유한 사람 : 
김연지

```

````python
# Sol 게시물 보여주기
class FacebookPost:
    
    def __init__(
        self, poster, time, public,
    content, wholiked, whocommented,
    whoshared):
        self.poster = poster
        self.time = time
        self.content = content
        self.wholiked = wholiked
        self.whocommented = whocommented
        self.whoshared = whoshared
    
    def show(self):
        linefeed = '\n'  # f-string 표현식에서 백슬래시(\) 못 써서 이렇게 썼어요.
        print(f'게시자 : {self.poster}\n')
        print(f'올린 시간 : {self.time}\n')
        print(f'내용 : \n{self.content}\n')
        print(f'좋아요 한 사람 : \n{linefeed.join(self.wholiked)}\n')
        print(f'댓글 남긴 사람 : \n{linefeed.join(self.whocommented)}\n')
        print(f'공유한 사람 : \n{linefeed.join(self.whoshared)}\n')
````

````python
post = FacebookPost(
    poster='Bad-Mouth',
    time='2시간',
    public=True,
    content='오토바이 : ????.gif\n존나 황당하겠다 ㄹㅇㅋㅋㅋㅋ',
    wholiked=['김민석', '유재건', '서용훈', '박성준'],
    whocommented=['김성재', '조영준', '우수한'],
    whoshared=['김연지'])
post.show()
````
