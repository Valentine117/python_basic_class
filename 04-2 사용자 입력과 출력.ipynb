우리들이 사용하는 대부분의 완성된 프로그램은 사용자의 입력에 따라 그에 맞는 출력을 내보낸다. 대표적인 예로 게시판에 글을 작성한 후 "확인" 버튼을 눌러야만(입력) 우리가 작성한 글이 게시판에 올라가는(출력) 것을 들 수 있다.

### 프롬프트를 띄워서 사용자 입력 받기

---
<br>

```python

```
위와 같이 입력하면 입력 받기 전에 사용자에게 질문을 띄울 수 있다.  

```python
>>> number = input("숫자를 입력하세요: ")

```
질문이 보인다는 것 외에 다른게 없다. 그냥 평소대로 입력하면 된다.  

```python
>>> number = input("숫자를 입력하세요: ")
숫자를 입력하세요: 3
>>> print(number)
3
>>>
```

````python
number = input("숫자를 입력하세요: ")
print(number)
````

#### 따옴표로 둘러싸인 문자열은 + 연산과 동일하다.

---
<br>

```python
>>> print("life" "is" "too short") # ①
lifeistoo short

lifeistoo short
```
위와 아래는 완전히 똑같다.  
*책에는 큰 따옴표라고 나와 있는데 작은 따옴표도 된다!*

## [문제 1] 두 수의 합은?

다음은 두 개의 숫자를 입력받아 더하여 리턴해 주는 프로그램이다.

```python

input2 = input("두번째 숫자를 입력하세요:")

total = input1 + input2
print("두 수의 합은 %s 입니다" % total)
```

