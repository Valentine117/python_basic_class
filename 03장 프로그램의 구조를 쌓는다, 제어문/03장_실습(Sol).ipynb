# 시험 성적
시험 점수($0 \le N \le 100$)을 입력받아 90 ~ 100점은 A, 80 ~ 89점은 B, 70 ~ 79점은 C, 60 ~ 69점은 D, 나머지 점수는 F를 출력하는 코드를 작성하여라.  
<br>

### hint
- `if` / `elif` / `else`를 이용해보자!

```python
# Sol 시험성적
score = int(input('score : '))
if 90 <= score <= 100:
    print('A')
elif 80 <= score:
    print('B')
elif 70 <= score:
    print('C')
elif 60 <= score:
    print('D')
else:
    print('F')
```

# 정렬
주어진 배열을 정렬하는 코드를 작성하여라.  
<br>

### hint
- `sorted` 사용금지
- 겹치는 숫자는 없다.
- `arr`를 수정한다.
- 배열의 원소는 모두 `int`이다.
- 오름차순으로 정렬한다.
- 정렬된 `arr`를 출력한다.

```python
arr = [34, 1, 8, 4, 17, 99, 6, 5, 26, 7, 10]
```

```python
# Sol 정렬
cnt = -1  # count의 줄임말, -1은 0이 아닌 아무 숫자나 넣은 것
while cnt != 0:
    cnt = 0
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            arr[i], arr[i+1] = arr[i+1], arr[i]
            cnt += 1
print(arr)
```

# 해피 버스데이~ (★)
2019년 1월 1일은 화요일이다. 1996년 5월 1일은 무슨요일이었을지 for문으로 구해보자!  
<br>

### hint
- 4의 배수 년도는 윤년으로 한다.
- 4의 배수지만 100의 배수인 년도는 평년(윤년이 아님)으로 한다.
- 100의 배수지만 400의 배수인 년도는 윤년으로 한다.
- 오늘부터 며칠 전인지 계산해보자!

```python
# Sol 해피 버스데이~
days_of_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
days = [x + '요일' for x in '월화수목금토일']
total_days = 0
for year in range(1996, 2019):
    if year == 1996:
        total_days += sum(days_of_month[4:])  # 4라고 써있지만 5월부터임
        continue
    total_days += sum(days_of_month)
    if year % 4 == 0:
        total_days += 1
    if year % 100 == 0:
        total_days -= 1
    if year % 400 == 0:
        total_days += 1
print(days[1 - (total_days % 7)])  # 1은 화요일!
```

# 별 찍기 5
N을 입력받아 아래와 같은 N층의 나무를 출력하는 코드를 작성해보자.  
<br>

### hint
- **`str.format`,  f-string 사용금지!**  
- 고등학교 때 배웠던 수열을 생각해보자!

```python
# Sol 별 찍기 5
N = int(input('N : '))
for i in range(N):
    # 별의 수 : 2i + 1
    # 너비 : 2N - 1
    # 양쪽 공백의 수 : (2N - 1)  - (2i + 1) = 2N - 2i - 2
    print(' ' * (N - i - 1) + '*' * (2 * i + 1) + ' ' * (N - i - 1))
```

# 소수 찾기 1

# 소수 찾기 2
2 이상의 자연수 N을 입력받아 N 미만의 소수를 모두 출력하는 코드를 작성하여라.  
<br>

### hint
- 입력하는 수는 무조건 자연수이니 다른 값이 들어올 걱정은 안 해도 된다.
- 중첩된 for 문을 이용해보자.

```python
# Sol 소수 찾기 2
from math import sqrt

N = int(input('N : '))
for i in range(2, N):
    cnt = 0
    for j in range(2, round(sqrt(N))):  # sqrt : square root, 왜 sqrt(N)까지만 탐색할까요?
        cnt += 1 if i % j == 0 else 0
    print(i) if cnt == 0 else None
```

# 소수 찾기 3
다음은 1798년 르장드르가 발표한 소수정리이다.

> 어떤 큰 자연수 $N$에 대하여 그 수가 소수일 확률은 ${1 \over \log N}$에 가깝다.  
여기서 $\log$는 자연로그이다.

**이는 어떤 큰 자연수 $N$ 이하의 소수의 개수가 ${N \over \log N}$에 가깝다는 것과 같다.**  
이를 증명해보자.

### hint
- 뭔가 주저리주저리 써있지만 **어려운 문제는 아니다.**
- 로그 계산은 `math`모듈의 `log`메소드를 이용한다.
- 코드 첫 줄에 `from math import log`를 입력하면 `log`함수를 사용할 수 있다.
- N은 십만 이상, 천만 이하의 큰 수를 입력한다. (그 이상 입력하면 오늘 안에 결과를 못 볼 수도 있다.)
- 큰 수를 입력하고 나면 계산 시간이 수 초에서 수 분 정도 걸린다. (본인 백만개 2분) 기다리면 된다.
- 출력 방식은 자유이다. 소수 개수와 ${N \over \log N}$이 비슷하다는 것을 보이면 된다.

```python
# Sol 소수 찾기 3
from math import sqrt, log
from IPython.core.display import clear_output  # 주피터에서 output을 지워주는 함수이다.

num_of_prime_numbers = 0

N = int(input('N : '))
for i in range(2, N+1):
    cnt = 0
    for j in range(2, round(sqrt(N+1))):
        if i % j == 0:
            cnt += 1
    if cnt == 0:
        num_of_prime_numbers += 1
    if i % 10000 == 0:
        clear_output()
        print(i)

N_logN = N / log(N)
print(f'# of prime numbers : {num_of_prime_numbers}, N/logN : {N_logN:.4f}')
```

# 자판기
`while`을 이용하여 자판기를 만들어보자.  
<br>
### hint
- 자연수 N을 받아 메뉴를 정한다.
- 자판기에는 코카콜라, 사이다, 닥터페퍼가 있다. (펩시는 먹지 않도록 한다.)
- 4를 입력하면 판매를 종료한다.

```python
# Sol 자판기
menu = ['코카콜라', '사이다', '닥터페퍼', '판매종료']
while True:
    command = int(input('\n'.join([f'{i+1}. {menu[i]}' for i in range(4)]) + '\ncommand : '))
    if command == 4:
        print('판매가 종료되었습니다.')
        break
    else:
        print(f'{menu[command-1]}를 뽑았습니다!')
```
