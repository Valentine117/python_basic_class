````python
f = open("doit/새파일.txt", 'w')
f.close()
````

파일을 생성하기 위해 우리는 `open`이라는 파이썬 내장 함수를 사용했다. `open` 함수는 다음과 같이 "파일 이름"( `"새파일.txt"` )과 "파일 열기 모드"( `'w'` )를 입력값으로 받고 결과값으로 파일 객체를 돌려준다.  
<br>
파일을 관리할 때 실수 방지와 호환을 위해 여러가지 모드를 설정할 수 있다.
> `파일 객체 = open(파일 이름, 파일 열기 모드)`

파일 열기 모드에는 다음과 같은 것들이 있다.

| 파일열기모드 |                        설명                         |
|------------|----------------------------------------------------|
|     r      |           읽기모드 - 파일을 읽기만 할 때 사용           |
|     w      |           쓰기모드 - 파일에 내용을 쓸 때 사용           |
|     a      | 추가모드 - 파일의 마지막에 새로운 내용을 추가 시킬 때 사용  |

쓰기모드 같은 경우에는 파일이 없으면 새로 만들고 있으면 안의 내용을 모두 지우고 연다.  
추가모드는 파일 내용을 지우지 않고 커서를 맨 끝으로 둔 상태로 열린다. 이 상태로 쓰면 파일 뒤에 내용을 덧붙일(append) 수 있다.
<br>
만약 새파일.txt라는 파일을 `doit`이라는 디렉터리에 생성하고 싶다면 다음과 같이 작성해야 한다.

```python
f = open("doit/새파일.txt", 'w')
f.close()
```

````python
f = open("doit/새파일.txt", 'w')
f.close()
````

위의 예에서 `f.close()`는 열려 있는 파일 객체를 닫아 주는 역할을 한다. 사실 이 문장은 생략해도 된다. 프로그램을 종료할 때 파이썬 프로그램이 열려 있는 파일의 객체를 자동으로 닫아주기 때문이다. 하지만 `close()`를 사용해서 열려 있는 파일을 직접 닫아 주는 것이 좋다.  
<br>
***쓰기모드로 열었던 파일을 닫지 않고 다시 사용하려고 하면 오류가 발생하기 때문이다.***

## <center>**파일을 쓰기 모드로 열어 출력값 적기**</center>

위에서는 파일을 열자마자 닫아서 아무 것도 쓰지 않았다. 이제 파일에 출력 내용을 써보자!
```python
f = open("doit/새파일.txt", 'w')
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```
아래 코드와 굉장히 비슷한 걸 알 수 있다.
```python
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    print(data, end='')
```
그냥 `print`가 `write`로 바뀌었고 파일 여닫는 코드가 추가됐을 뿐이다.  
정확히 말하자면 `print`는 콘솔(아래 뜨는 창을 그냥 콘솔이라 해요)을 통해 모니터에 출력하고 `write`는 그 내용을 파일에 출력하는 것이다.  
<br>
아래 코드를 실행해보자!

````python
f = open("doit/새파일.txt", 'w')
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
````


없던 파일이 새로 생성된 걸 볼 수 있다.
이걸 열어보면

아래 `print`를 이용한 코드와 출력 결과가 똑같은 걸 볼 수 있다.  
자꾸 강조하면 중요한거다.

````python
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    print(data, end='')
````

## <center>**프로그램 외부에 저장된 파일을 읽는 여러가지 방법**</center>

<br>

---
파일을 읽을 때 사용할 수 있는 파이썬 내장함수는 크게 세가지가 있다.
```python
readline()
readlines()
read()
```
세 함수 다 많이 쓰인다. 어떻게 쓰이는지 알아보자.

### readline 이용하기

---
<br>

```python
f = open("doit/새파일.txt", 'r')
line = f.readline()
print(line)
f.close()
```
출력 한 번만 해보면 뭐하는 함수인지 알 수 있다.

````python
f = open("doit/새파일.txt", 'r')
line = f.readline()
print(line)
f.close()
````

그렇다. 한 줄씩 읽어오는 함수이다.  
파일 전체를 읽어오고 싶으면 함수를 여러 번 호출하면 된다.
```python
f = open("doit/새파일.txt", 'r')
while True:
    line = f.readline()
    if not line: break
    print(line, end='')
f.close()
```
`f.readline()`은 더이상 읽을 라인이 없으면 `None`을 반환한다.  
`None`이 되면 `not None`이 `True`가 되면서 루프를 탈출한다.

````python
f = open("doit/새파일.txt", 'r')
while True:
    line = f.readline()
    if not line: break
    print(line, end='')
f.close()
````

그리고 이건 마치
```python
while True:
    data = input()
    if not data: break
    print(data)
```
와 비슷하다.  
책에서 `input`과 파일을 읽어오는 `read`가 비슷하다는 걸 자꾸 강조하는데 이 정도면 이해 됐으리라 본다.

### readlines 이용하기

---
<br>

```python
f = open("doit/새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```
`readlines`는 파일 내용을 한 번에 받아와 줄 단위로 리스트에 넣어 주는 함수이다.  

> ### 파이썬에서는 `f.close()`를 잊어도 됩니다!
파일을 사용해놓고 안 닫는 것은 프로그래밍에 익숙한 사람도 자주 하는 실수이다.  
파이썬에서는 이를 방지하기 위해 특별한 키워드를 제공한다.
```python
with open('doit/새파일.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        print(line)
```
위와 같이 `with`를 사용해 파일 객체( `f` )를 얻을 수 있고,  
`with` scope를 벗어나면 자동으로 `f.close()`를 실행해준다.  
이제 파일 입출력 코드는 이렇게 쓰자!

````python
with open("doit/새파일.txt", 'r') as f:
    lines = f.readlines()
lines
````

> ### `readline`과 `readlines`의 차이점
`readline`은 한 줄씩 읽어와서 메모리가 절약되지만 "n번째 줄"과 같이 특정 원소에 바로 접근(랜덤 액세스라 한다)이 불가능하다.  
`readlines`는 내용을 한 번에 받아와서 랜덤 액세스가 가능하지만 파일의 용량이 크면 메모리를 많이 잡아먹는다는 단점이 있다.

````python
for line in lines:
    print(line, end='')
````

### read 이용하기

---
<br>

```python
with open('doit/새파일.txt', 'r') as f:
    data = f.read()
    print(data)
```
`read`는 내용을 줄 상관없이 한 번에 받아온다. 보통 JSON파일을 받아올 때 많이 쓴다.  

> ### JSON이 몬가욤???
JavaScript Object Notation(제이슨)의 준말로 객체를 파일로 저장하는 방법 중에 하나에요.  
직관적이고 편해서 대유행을 타고 있습니다.  
지금 쓰고 있는 이 주피터 노트북도 사실 JSON형식으로 이루어져 있고 그걸 기반으로 구동되고 있어요.  
객체가 뭔지 JSON이 얼마나 편한지는 class를 배우면 알게 될거에요!

````python
with open('doit/새파일.txt', 'r') as f:
    data = f.read()
    print(data)
````

## <center>**파일에 새로운 내용 추가하기**</center>

파일에 뭔가 덧붙이고 싶다고 `'w'`로 열어버리면 파일의 내용이 모두 사라지게 된다. 이런 경우는 appending모드인 `'a'`로 열면 된다. 
```python
with open('doit/새파일.txt', 'a') as f:
    for i in range(11, 20):
        data = "%d번째 줄입니다.\n" % i
        f.write(data)
```
이 코드를 실행하고 파일을 확인해보자.


````python
with open('doit/새파일.txt', 'a') as f:
    for i in range(11, 20):
        data = "%d번째 줄입니다.\n" % i
        f.write(data)
````
