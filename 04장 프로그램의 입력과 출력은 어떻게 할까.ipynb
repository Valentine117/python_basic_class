# 사용자 입력과 출력
우리들이 사용하는 대부분의 완성된 프로그램은 사용자의 입력에 따라 그에 맞는 출력을 내보낸다. 대표적인 예로 게시판에 글을 작성한 후 "확인" 버튼을 눌러야만(입력) 우리가 작성한 글이 게시판에 올라가는(출력) 것을 들 수 있다.

> 사용자 입력 → 처리(프로그램, 함수 등) → 출력

```python
>>> for i in range(10):
>>>    print(i)
```
